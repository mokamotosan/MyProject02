# Confirmatory Analyses

## 01: a one-tailed independent two-sample t-test
The author had planned:
- sample size: 51 for each group
- alpha: 0.05
- effect size: 0.5
- power: 0.8

```python
groupH = df[df['IOS_group']=='HighIOS']
groupL = df[df['IOS_group']=='LowIOS']
```

### Overlapping Magnitude

```python
stats.ttest_ind(groupH['Overlap_score'], groupL['Overlap_score'], equal_var=False) # perform Welch's t-test without assuming equal population variance.
```

### Euclidean Distance

```python
stats.ttest_ind(groupH['euclidean_distance'], groupL['euclidean_distance'], equal_var=False) # perform Welch's t-test without assuming equal population variance.
```

### Manhattan Distance

```python
stats.ttest_ind(groupH['manhattan_distance'], groupL['manhattan_distance'], equal_var=False) # perform Welch's t-test without assuming equal population variance.
```

### Cosine Similarity

```python
stats.ttest_ind(groupH['cosine_similarity'], groupL['cosine_similarity'], equal_var=False) # perform Welch's t-test without assuming equal population variance.
```

### Post hoc Power Analysis

```python
print('Num. of group H: ' + str(len(groupH)))
print('Num. of group L: ' + str(len(groupL)))
```

Achived Power: 0.85

## 02: Spearman's rank correlation coefficient between the IOS scores and the overlapping scores
The author had planned:
- sample size: 119, including participants who rated on the IOS scale as 4
- alpha: 0.05
- effect size: 0.3
- power: 0.93
- two-tailed


### Overlapping score

```python
iosscores = df['IOS_score']
overlapscores = df['Overlap_score']
```

```python
res = stats.spearmanr(iosscores, overlapscores)
res
```

### Euclidean Distance

```python
euclidean_distance = df['euclidean_distance']
res = stats.spearmanr(iosscores, euclidean_distance)
res
```

### Manhattan Distance

```python
manhattan_distance = df['manhattan_distance']
res = stats.spearmanr(iosscores, manhattan_distance)
res
```

### Cosine Similarity

```python
cosine_similarity = df['cosine_similarity']
res = stats.spearmanr(iosscores, cosine_similarity)
res
```

### Post hoc Power Analysis

```python
print('Num. of Sumple of the IOS socre: ' + str(len(iosscores)))
print('Num. of Sumple of the IOS socre: ' + str(len(overlapscores)))
```

Achieved power: 0.95
